```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

import auto_point_detection

from calibration_utils import load_thermal_image, load_rgb_image

def test_calibration(thermal_path, rgb_path, is_hydra=True):
    print(rgb_path)
    thermal_1 = load_thermal_image(thermal_path)

    rgb_1 = load_rgb_image(rgb_path)
    if rgb_1.ndim == 3:
        rgb_1 = cv2.cvtColor(rgb_1, cv2.COLOR_RGB2GRAY)

    _, thermal_raw, thermal_thresh = auto_point_detection.find_calibration_points_on_heatmap(thermal_1, is_hydra=is_hydra)
    _, rgb_raw, rgb_thresh = auto_point_detection.find_calibration_points_on_rgb_photo(rgb_1)

    fig1, axs1 = plt.subplots(2, 2, figsize=(10,15))

    fig1.tight_layout()

    axs1[0][0].imshow(thermal_raw, cmap='gray')

    axs1[0][1].imshow(thermal_thresh, cmap='gray')

    axs1[1][0].imshow(rgb_raw, cmap='gray')

    axs1[1][1].imshow(rgb_thresh, cmap='gray')

    plt.show()


test_calibration('wip_dataset/thermal/hydra_1.npy', 'wip_dataset/rgb/hydra_1.png', is_hydra=True)
test_calibration('wip_dataset/thermal/hydra_3.npy', 'wip_dataset/rgb/hydra_3.png', is_hydra=True)
test_calibration('wip_dataset/thermal/hydra_4.npy', 'wip_dataset/rgb/hydra_4.png', is_hydra=True)
test_calibration('wip_dataset/thermal/hydra_5.npy', 'wip_dataset/rgb/hydra_5.png', is_hydra=True)
test_calibration('wip_dataset/thermal/hydra_6.npy', 'wip_dataset/rgb/hydra_6.png', is_hydra=True)
test_calibration('wip_dataset/thermal/hub_1.npy', 'wip_dataset/rgb/hub_1.png', is_hydra=False)
test_calibration('wip_dataset/thermal/hub_2.npy', 'wip_dataset/rgb/hub_2.png', is_hydra=False)
test_calibration('wip_dataset/thermal/hub_3.npy', 'wip_dataset/rgb/hub_3.png', is_hydra=False)
```
